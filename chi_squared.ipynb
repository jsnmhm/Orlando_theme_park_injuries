```python
import pandas as pd
import re
from scipy.stats import chi2_contingency
```

```python
def get_age_category(age_gender):
    age = float(re.findall(r'\d+', age_gender)[0])  # Extract the numeric age from the string
    if 1 <= age <= 12:
        return "Children"
    elif 13 <= age <= 17:
        return "Adolescents"
    elif 18 <= age <= 34:
        return "Younger Adults"
    elif 35 <= age <= 64:
        return "Adults"
    elif age >= 65:
        return "Older adults"
    else:
        return "Unknown"  # Handle any other cases if needed
```

```python
df = pd.read_csv("theme_park_injuries_ride_type.csv")
df["age_category"] = df["age_gender"].apply(get_age_category)
df.head()
cross_table = pd.crosstab(df["age_category"], df["type"])
cross_table.head()
```

```python
chi2, p_value, dof, expected = chi2_contingency(cross_table)

# Output the results
print("Chi-Squared test statistic:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(pd.DataFrame(expected, index=cross_table.index, columns=cross_table.columns))
```

```python
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant association between age group and ride type.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between age group and ride type.")
```
