```python
# Calculate the number of incidents for each company
incident_counts = df['Company'].value_counts()

# Print the number of incidents for each company
print("Incident Counts:")
print(incident_counts)

# Separate the data for each company into two samples
company1_data = df[df['Company'] == 'Disney World']
company2_data = df[df['Company'] == 'Universal']

# Perform Independent Samples T-Test
t_statistic, p_value = stats.ttest_ind(company1_data.index, company2_data.index)

# Print the results
print("Independent Samples T-Test Results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Check if the p-value is less than a significance level (e.g., 0.05) to determine significance
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference in the amount of incidents between the two companies.")
else:
    print("There is no statistically significant difference in the amount of incidents between the two companies.")
```

```python
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest


# Clean the 'Company' column by removing leading/trailing whitespaces
df['Company'] = df['Company'].str.strip()

# Calculate the number of incidents for each company
incident_counts = df['Company'].value_counts()

# Total number of visitors per year for Disney World and Universal
visitors_disney = 58_000_000  # 58 million
visitors_universal = 22_000_000  # 22 million

# Calculate the incident rates for each company
incident_rates_disney = incident_counts['Disney World'] / visitors_disney
incident_rates_universal = incident_counts['Universal'] / visitors_universal

# Perform the two-sample Z-test for proportions
z_statistic, p_value = proportions_ztest([incident_counts['Disney World'], incident_counts['Universal']],
                                         [visitors_disney, visitors_universal])

# Print the results
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)

# Check if the p-value is less than a significance level (e.g., 0.05) to determine significance
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference in incident rates between the two companies.")
else:
    print("There is no statistically significant difference in incident rates between the two companies.")
```

```python
orl_lat = 28.5384
orl_lon = -81.3789
radius = 75000

params = {
    "apiKey":api_key,
    "filter":f"circle:{orl_lon},{orl_lat},{radius}"
}

#add empty columns for lat, lon
rides["lat"] = None
rides["lon"] = None

    
base_url = "https://api.geoapify.com/v1/geocode/search"

for index, row in rides.iterrows():
    ride_name = row["Ride_name"]
    params["text"] = ride_name
    ride_data = requests.get(base_url, params=params).json()
    try:
        rides.loc[index, "lat"] = ride_data["features"][0]["properties"]["lat"]
        rides.loc[index, "lon"] = ride_data["features"][0]["properties"]["lon"]
    except IndexError:
        print("No record found")
        
rides.head()
```

```python
def plot_mapping_data(mapping_data, filter_column=None, 
                      filter_value=None, scale=5, save_file=None, title="title"):
    
    if filter_column is not None and filter_value is not None:
        filtered_data = mapping_data[mapping_data[filter_column] == filter_value]
    else:
        filtered_data = mapping_data
    
    map_plot = filtered_data.hvplot.points("lon",
                                     "lat",
                                     legend=True,
                                     color="Ride_name",
                                     tiles="CartoLight",
                                     size="incident_count",
                                     scale=scale,
                                     width=1000,
                                     height=750,
                                     alpha=0.75,
                                     title=title)
    
    hvplot.save(map_plot, f"maps\{title}.png")
    
    return map_plot
```

```python
import pandas as pd
import re
from scipy.stats import chi2_contingency
```

```python
def get_age_category(age_gender):
    age = float(re.findall(r'\d+', age_gender)[0])  # Extract the numeric age from the string
    if 1 <= age <= 12:
        return "Children"
    elif 13 <= age <= 17:
        return "Adolescents"
    elif 18 <= age <= 34:
        return "Younger Adults"
    elif 35 <= age <= 64:
        return "Adults"
    elif age >= 65:
        return "Older adults"
    else:
        return "Unknown"  # Handle any other cases if needed
```

```python
df = pd.read_csv("theme_park_injuries_ride_type.csv")
df["age_category"] = df["age_gender"].apply(get_age_category)
df.head()
cross_table = pd.crosstab(df["age_category"], df["type"])
cross_table.head()
```

```python
chi2, p_value, dof, expected = chi2_contingency(cross_table)

# Output the results
print("Chi-Squared test statistic:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(pd.DataFrame(expected, index=cross_table.index, columns=cross_table.columns))
```

```python
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant association between age group and ride type.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between age group and ride type.")
```
